```python

A = [1, 4, 3, 5, 2, 4, -33, 283, 0, -1]



def maxmin( A, start, end ) :

    if len(A) == 1: 
        return ( A[start], A[end] )
    
    if (end-start) == 1 :  
        if A[start] <= A[end] :
            return ( A[start], A[end] )
        else :
            return ( A[end], A[start] )

    if len(A) % 2 == 0:
        middle = (len(A))/2
    else:
        middle = (len(A)+1)/2

    set_1 = maxmin( A, start, middle )
    set_2 = maxmin( A, middle+1, end )

    maxim = max( set_1[1], set_2[1] )
    minim = min( set_1[0], set_2[0] )
    

    return (maxim, minim)

output = maxmin(A, 0, len(A)-1)
print(output)
```
