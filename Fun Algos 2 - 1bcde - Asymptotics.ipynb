```python
# 1b
# Michael Lukiman - Fun Algos

import scipy, numpy as np, matplotlib as mpl, pylab as pl

x = np.linspace(6,200,100)
y = np.log(x**3-(30*x))
z = np.log(x**2)

def compare_funcs():
    pl.plot(x,y)
    pl.plot(x,z,'co') # second functiog g(n) in blue dots
    pl.show()
    
compare_funcs()
```
